```python
query ='''
SELECT * FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
'''

# retrieve data from query and convert it into dataframe
df = client.query(query).to_dataframe()

# print the dataframe
len([col for col in df.columns if df[col].dtype == 'object'])
'''
Explanation
[col for col in df.columns if df[col].dtype == 'object']

This part is a list comprehension in Python. It's a compact way to create a list based on conditions.
for col in df.columns: Here, we’re iterating over each column name in df.columns. This gives us each column one by one.
df[col].dtype: This checks the data type (dtype) of the current column col.
df[col].dtype == 'object': In Pandas, columns with dtype == 'object' usually contain strings (although sometimes they can contain mixed data types too). This condition checks if the current column is of type object.
The if statement filters only those columns for which df[col].dtype is 'object'.
So, [col for col in df.columns if df[col].dtype == 'object'] creates a list of column names that have a data type of 'object'.
len(...)

Finally, we pass this list to the len() function. len() returns the number of items in the list, which is the count of columns that have a string (or object) data type.
Summary
This code counts the number of columns in df that are of type 'object', which usually means they contain strings.
'''
```
